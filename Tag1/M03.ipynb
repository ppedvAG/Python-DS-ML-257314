## Serie
Ein Numpy Array mit extra Eigenschaften
- Name
- Benannter Index
- ...
Auf das unterliegen Numpy Array kann immer zugegriffen werden

```python
e=pd.Series([8.6, 87.9, 0.23, 9]) # diese Zahlen alleine sagen viel aus
```

### Name

```python
e.name = "Einwohnerzahlen (Millionen)"
```

```python
e
```

### Index

```python
e.index = ["CH", "DE", "LI", "AT"]
```

```python
e
```

```python
e["DE"]
```

```python
e["DE":"AT"]
```

```python
e.values # Das unterliegene numpy Array
```

#### Vektorisierung wie bei Arrays

```python
e>5
```

```python
e[e>5]
```

```python
e[e>e.mean()]
```
