##### Index setzen
set_index("Spaltenname) ODER index_col direkt bei read_csv

__WICHTIG__: Jeder Befehl erzeugt immer nur eine Kopie des Datensets
Zwei Optionen:
- Bestehende Variable überschreiben
- inplace

```python
# data = data.set_index("Country (or dependency)")
```

```python
data.set_index("Country (or dependency)", inplace=True)
```

```python
data
```

```python
data.index.name = "Country"
```

```python
data
```

##### Spalten anpassen
- entfernen mit drop
- umbenennen mit rename (rename benötigt ein Dictionary mit den colums = Parameter)


```python
data.drop(columns=["#"], inplace=True)
```

```python
data
```

```python
data.rename(columns= {
    "Population(2020)": "Population",
    "Density(P/Km²)": "Density",
    "Land Area(Km²)": "LandArea",
    "Migrants(net)": "Migrants",
    "Fert.Rate": "FertRate",
    "Med.Age": "MedAge",
    "UrbanPop %": "UrbanPopPct",
    "WorldShare": "WorldSharePct"
}, inplace = True)
```

```python
data
```

##### Daten speichern


```python
data.to_csv("data/PopulationDataNearlyDone.csv")
```

###### Sortieren
- sort_index()
- sort_values(["S1", "S2])
Alle Sortierungen können auch mit ascending =True/False angegeben werden

```python
data.sort_index()
```

```python
data.sort_index(ascending = False)
```

```python
data.sort_values("LandArea", ascending = False)
```

```python
data.sort_values(["LandArea", "Population"], ascending = False)
```

```python
data.sort_values(["LandArea", "Population"], ascending = [False, True])
```

##### Auf Daten zugreifen
Drei Möglichkeiten
- Index: Zugriff auf Zeilen/Spalten Kombinationen
- loc: Mehrere Spalten
- iloc: Wie loc, aber nur mit Zahlen

```python
data["Population"] # Eine Spalte oder Zeile eines Dictionarys ist eine Serie
```

```python
data["China":"China"]
```

```python
data["China":"Russia"] # meherere Zeilen
```

```python
data["Density"]["India"] #Einzelne Werte
```

```python
data["Population"]["China":"Russia"] #Mehrere Zeilen einer Spalte
```

```python
data.loc["China":"Russia", "Population":"Density"] #Mehrere Zeilen einer Spalte
```

```python
data.loc["China":"Russia", ("Population","Density")] # Einzelne Spalten selektieren
```

```python
data.loc[("China","Russia"), :] # Einzelne Zeilen selektieren
```

```python
data.iloc[10:20, 0:3]
```
