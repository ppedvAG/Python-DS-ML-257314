```python
b = np.random.randint(10, size=100)
c = np.random.randint(10, size=100)
```

```python
b == c
```

```python
# Gleiche Werte
b[b == c]
```

```python
# Stellen
np.arange(100)[b == c]
```

```python
# Beides gleichzeitig
maske = b == c
np.vstack((np.arange(100)[maske], b[maske], c[maske]))
```
