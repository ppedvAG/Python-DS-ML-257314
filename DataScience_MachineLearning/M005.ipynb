```python
population = pd.read_csv("data/PopulationDataDone.csv", index_col=0)[:10]
```

```python
plt.pie(population["WorldSharePct"], labels=population.index, rotatelabels=True, explode=np.ones(10)*0.02)
```

```python
plt.figure(figsize=(12,4))
plt.bar(population.index, population["WorldSharePct"])
plt.xticks(rotation=70)
```

## Globaler vs Objektorientierter Modus
Globaler Modus: Befehle ausführen über plt, mit plt.show() die Konfiguration abschließen

Objektorientierter Modus: Figure Objekte und Axis Objekte werden dediziert angesprochen und dadurch konfiguriert

```python
plt.plot(population["Population"])
plt.grid(alpha=0.3)
plt.savefig("Test2.png")
plt.show() # Schließt die Konfiguration des Plots ab und zeigt ihn an

plt.plot(population["WorldSharePct"])
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(population["Population"])
plt.grid(alpha=0.3)
plt.show()
```

```python
fig, ((ax1_1, ax1_2), (ax2_1, ax2_2)) = plt.subplots(2,2)
ax1_1.plot(np.random.random(100))
ax1_2.pie(population["WorldSharePct"])
ax2_1.bar(population.index, population["Population"])
ax2_2.hist(np.random.randint(100, size=100), bins=20)
fig.savefig("Test.png")
```
